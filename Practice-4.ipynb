```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import arange
from sklearn.linear_model import LassoCV,Lasso
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# Reading the file
```

```python
df=pd.read_excel('D:\SEM 4\Supervised Machine Learning Lab\\Lasso and Ridge Regression dataset.xlsx')
```

```python
# Dividing the dataset
```

```python
X=df[['X1','X2','X3','X4','X5','X6','X7','X8']]
Y1=df['Y1']
Y2=df['Y2']
```

```python
X
```

```python
Y1
```

```python
Y2
```

```python
X_train,X_test,Y1_train,Y1_test=train_test_split(X,Y1,test_size=0.2,random_state=42)
_,_,Y2_train,Y2_test=train_test_split(X,Y2,test_size=0.2,random_state=42)
```

```python
# Lasso Regression
```

```python
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
lasso_model = LassoCV(alphas=np.arange(0, 1, 0.01), cv=cv)
```

```python
lasso_model.fit(X_train, Y1_train)
```

```python
optimal_alpha_Y1 = lasso_model.alpha_
print("Optimal Alpha for Y1:", optimal_alpha_Y1)
```

```python
Y1_pred = lasso_model.predict(X_test)
mse_Y1 = mean_squared_error(Y1_test, Y1_pred)
rmse_Y1 = np.sqrt(mse_Y1)
print("MSE for Y1:", mse_Y1)
print("RMSE for Y1:",rmse_Y1)
r2_Y1 = r2_score(Y1_test, Y1_pred)
print("R-squared Score for Y1:", r2_Y1)
```

```python
lasso_model.fit(X_train, Y2_train)
```

```python
optimal_alpha_Y2 = lasso_model.alpha_
print("Optimal Alpha for Y2:", optimal_alpha_Y2)
```

```python
Y2_pred = lasso_model.predict(X_test)
mse_Y2 = mean_squared_error(Y2_test, Y2_pred)
rmse_Y2 = np.sqrt(mse_Y2)
print("MSE for Y2:", mse_Y2)
print("RMSE for Y1:",rmse_Y1)
r2_Y2 = r2_score(Y2_test, Y2_pred)
print("R-squared Score for Y2:", r2_Y2)
```

```python
# Ridge regression
```

```python
ridge_model = RidgeCV(alphas=np.arange(0.01, 1, 0.01), cv=cv)
```

```python
ridge_model.fit(X_train, Y1_train)
```

```python
optimal_alpha_Y1_ridge = ridge_model.alpha_
print("Optimal Alpha for Y1 (Ridge):", optimal_alpha_Y1_ridge)
```

```python
Y1_pred_ridge = ridge_model.predict(X_test)
mse_Y1_ridge = mean_squared_error(Y1_test, Y1_pred_ridge)
rmse_Y1_ridge = np.sqrt(mse_Y1_ridge)
print("MSE for Y1 (Ridge):", mse_Y1_ridge)
print("RMSE for Y1:",rmse_Y1_ridge)
r2_Y1_ridge = r2_score(Y1_test, Y1_pred_ridge)
print("R-squared Score for Y1 (Ridge):", r2_Y1_ridge)
```

```python
ridge_model.fit(X_train, Y2_train)
```

```python
optimal_alpha_Y2_ridge = ridge_model.alpha_
print("Optimal Alpha for Y2 (Ridge):", optimal_alpha_Y2_ridge)
```

```python
Y2_pred_ridge = ridge_model.predict(X_test)
mse_Y2_ridge = mean_squared_error(Y2_test, Y2_pred_ridge)
rmse_Y2_ridge = np.sqrt(mse_Y2_ridge)
print("MSE for Y2 (Ridge):", mse_Y2_ridge)
print("RMSE for Y2:",rmse_Y2_ridge)
r2_Y2_ridge = r2_score(Y2_test, Y2_pred_ridge)
print("R-squared Score for Y2 (Ridge):", r2_Y2_ridge)
```
