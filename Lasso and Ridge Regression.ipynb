```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import arange
from sklearn.linear_model import LassoCV # CV is cross validation
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import RepeatedKFold
```

```python
url="https://raw.githubusercontent.com/Statology/Python-Guides/main/mtcars.csv"
df=pd.read_csv(url)
```

```python
df = df[['mpg','wt','drat','qsec','hp']]
```

```python
df.head(10)
```

```python
X=df[['mpg','wt','drat','qsec']]
X
```

```python
y=df['hp']
y
```

```python
# LassoCV() then we use RepeatedKFold() to find the optimal lambda value for the penalty term
# We divide data into folds and also repeat the process n number of times
```

```python
cv=RepeatedKFold(n_splits=10,n_repeats=3,random_state=1) # to divide the data into 10 splits and then 3 times we repeat the dividing into 10 splits
```

```python
model=LassoCV(alphas=arange(0,1,0.01),cv=cv) #alphas is used to determine the lambda of the penalty term
model.fit(X,y)

# Keeps adding 0.01,0.02,... (Leads to change in slope) and finds the MSE and chooses the term that gives minimum MSE
```

```python
model.alpha_ #value which gives the least MSE value
```

```python
new=[24,2.5,3.5,18.5]
```

```python
model.predict([new])
```

```python
model1=RidgeCV(alphas=arange(0.01,1,0.01),cv=cv) #alphas is used to determine the lambda of the penalty term
model1.fit(X,y)
```
